```
for file in os.listdir('/content/Object-classification-using-VGG16/images'):
  print(file)
  full_path ='/content/Object-classification-using-VGG16/images/' + file
  img= image.load_img(full_path ,target_size= (224,224)) #This loads an image and resizes the image to (224, 224)
  img= image.img_to_array(img) #The img_to_array() function adds channels: x.shape = (224, 224, 3) for RGB and (224, 224, 1) for gray image
  img= np.expand_dims(img, axis=0) #expand_dims() is used to add the number of images: x.shape = (1, 224, 224, 3)
  img = preprocess_input(img) #preprocess_input subtracts the mean RGB channels of the imagenet dataset. This is because the model you are using has been trained on a different dataset: x.shape is still (1, 224, 224, 3)
  y_pred = model.predict(img)
  lable=decode_predictions(y_pred,top=1)
  print(lable)
  print()
```
